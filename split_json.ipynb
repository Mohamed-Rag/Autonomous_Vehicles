```python
import os
import json
import shutil
from tqdm import tqdm
```

```python

# ==== Change paths ====
labels_dir = "/home/ragab/Desktop/summer/python/depi/project/bdd100k_labels./100k/train"        # original JSON folder
output_dir = "/home/ragab/Desktop/summer/python/depi/project/final_dataset/train/labels"       # where to save split JSONs

time_categories = ["daytime", "night", "dawn/dusk"]

# Create output folders
for cat in time_categories:
    os.makedirs(os.path.join(output_dir, cat), exist_ok=True)

# Split JSONs
for label_file in tqdm(os.listdir(labels_dir), desc="Splitting JSONs"):
    if not label_file.endswith(".json"):
        continue

    label_path = os.path.join(labels_dir, label_file)
    with open(label_path, "r") as f:
        data = json.load(f)

    time_of_day = data["attributes"].get("timeofday", None)

    if time_of_day in time_categories:
        dst_json = os.path.join(output_dir, time_of_day, label_file)
        shutil.copy(label_path, dst_json)

print("✅ JSON files have been split into daytime, night, dawn/dusk")
```
