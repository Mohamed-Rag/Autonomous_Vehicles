```python
import os
import json
import shutil
from tqdm import tqdm
```

```python
# Train
images_dir = "/home/ragab/Desktop/summer/python/depi/project/bdd100k_images_100k/100k/train"
labels_dir = "/home/ragab/Desktop/summer/python/depi/project/bdd100k_labels./100k/train"
output_dir = "/home/ragab/Desktop/summer/python/depi/project/final_dataset/train"

# Create output folders
time_categories = ["daytime", "night", "dawn/dusk"]
for cat in time_categories:
    os.makedirs(os.path.join(output_dir, cat), exist_ok=True)

# Collect all JSON files
json_files = [f for f in os.listdir(labels_dir) if f.endswith(".json")]

# Track which images have labels
labeled_images = set()

# Process JSON files
for label_file in tqdm(json_files, desc="Processing JSONs"):
    label_path = os.path.join(labels_dir, label_file)
    with open(label_path, "r") as f:
        data = json.load(f)

    # Get image name
    img_name = data["name"]
    if not img_name.endswith(".jpg"):
        img_name += ".jpg"

    time_of_day = data["attributes"].get("timeofday", None)

    if time_of_day in time_categories:
        src_img = os.path.join(images_dir, img_name)
        dst_img = os.path.join(output_dir, time_of_day, img_name)

        if os.path.exists(src_img):
            shutil.copy(src_img, dst_img)
            labeled_images.add(img_name)


```

```python
# Remove images without labels 
for img_file in tqdm(os.listdir(images_dir), desc="Cleaning unlabeled images"):
    if img_file.endswith(".jpg") and img_file not in labeled_images:
        os.remove(os.path.join(images_dir, img_file))
```
